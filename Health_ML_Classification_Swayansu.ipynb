# Building Reusable Function

```python
def summarize_classification(y_test,y_pred):
  acc = accuracy_score(y_test,y_pred,normalize=True)
  num_acc = accuracy_score(y_test,y_pred,normalize=False)
  prec = precision_score(y_test,y_pred)
  recall = recall_score(y_test,y_pred)
  return{
      'accuracy':acc,
      'precision':prec,
      'recall':recall,
      'accuracy_count':num_acc
  }
```

```python
def build_model(classifier_fn,name_of_y_col,names_of_x_cols,dataset,test_frac=0.2):
  X=dataset[names_of_x_cols]
  Y=dataset[name_of_y_col]
  x_train, x_test, y_train, y_test = train_test_split( X, Y, test_size=test_frac)
  model = classifier_fn(x_train,y_train)
  y_pred = model.predict(x_test)
  y_pred_train = model.predict(x_train)
  train_summary = summarize_classification(y_train,y_pred_train)
  test_summary = summarize_classification(y_test,y_pred)
  pred_results = pd.DataFrame({'y_test':y_test,
                               'y_pred':y_pred})
  model_crosstab = pd.crosstab(pred_results.y_pred,pred_results.y_test)
  return{'training':train_summary,
         'test':test_summary,
         'confusion_matrix': model_crosstab}
```

```python
def compare_results():
  for key in result_dict:
    print('Classification: ',key)

    print()
    print('Training Data')
    for score in result_dict[key]['training']:
      print(score, result_dict[key]['training'][score])

    print()
    print('Test Data')
    for score in result_dict[key]['test']:
      print(score, result_dict[key]['test'][score])

    print()
```

# Experimenting Different classification Algorithms

```python
def logistic_fn(x_train,y_train):
  model = LogisticRegression(solver='liblinear')
  model.fit(x_train,y_train)

  return model
```

```python
result_dict['health - logistic'] = build_model(logistic_fn,'DEATH_EVENT',Features,df)
compare_results()
```

```python
def Linear_discrim_fn(x_train,y_train,solver='svd'):
  model = LinearDiscriminantAnalysis(solver=solver)
  model.fit(x_train,y_train)

  return model
```

```python
result_dict['health - Linear_discriminant'] = build_model(Linear_discrim_fn,'DEATH_EVENT',Features,df)
compare_results()
```

```python
def Quad_discrim_fn(x_train,y_train):
  model = QuadraticDiscriminantAnalysis()
  model.fit(x_train,y_train)

  return model
```

```python
result_dict['health - Quadtratic_discriminant'] = build_model(Quad_discrim_fn,'DEATH_EVENT',Features,df)
compare_results()
```

```python
def sgd_fn(x_train,y_train,max_iter=10000,tol=1e-3):
  model = SGDClassifier(max_iter=max_iter,tol=tol)
  model.fit(x_train,y_train)

  return model
```

```python
result_dict['health - Stochastic_Gradient_Descent'] = build_model(sgd_fn,'DEATH_EVENT',Features,df)
compare_results()
```

```python
def linear_svc_fn(x_train,y_train,C=1,max_iter=10000,tol=1e-3):
  model = LinearSVC(C=C,max_iter=max_iter,tol=tol,dual=False)
  model.fit(x_train,y_train)

  return model
```

```python
result_dict['health - SVC'] = build_model(linear_svc_fn,'DEATH_EVENT',Features,df)
compare_results()
```

```python
def Radius_neigh_fn(x_train,y_train,radius = 1000.0):
  model = RadiusNeighborsClassifier(radius=radius)
  model.fit(x_train,y_train)

  return model
```

```python
result_dict['health - Radius_neighbour'] = build_model(Radius_neigh_fn,'DEATH_EVENT',Features,df)
compare_results()
```

```python
def Decision_tree_fn(x_train,y_train,max_depth=None,max_features=None):
  model = DecisionTreeClassifier(max_depth=max_depth,max_features=max_features)
  model.fit(x_train,y_train)

  return model
```

```python
result_dict['health - Decision_Tree'] = build_model(Decision_tree_fn,'DEATH_EVENT',Features,df)
compare_results()
```

```python
def Naive_bayes_fn(x_train,y_train,priors=None):
  model = GaussianNB(priors=priors)
  model.fit(x_train,y_train)

  return model
```

```python
result_dict['health - Naive_Bayes'] = build_model(Naive_bayes_fn,'DEATH_EVENT',Features,df)
compare_results()
```
